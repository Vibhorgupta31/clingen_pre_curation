```python
# Library imports
import datetime
from dotenv import load_dotenv
from openai import OpenAI
import os
import requests
```

```python
load_dotenv()
```

```python
try:
    with open("Input/prompt.txt","r") as fh:
        system_prompt = fh.read()
except FileNotFoundError:
    print("Input/prompt.txt not found")
    system_prompt = ''
```

```python
try:
    with open("Input/RET.txt","r") as fh:
        user_input = fh.read()
except FileNotFoundError:
    print("Input/RET.txt not found")
    user_input = ''
```

```python
client = OpenAI()

response = client.responses.create(
    model="gpt-5",
    reasoning={"effort":"low"},
    input=[
        {
            "role":"system",
            "content":system_prompt
        },
        {
        "role":"user",
        "content":user_input
        }
    ]            
)
```

```python
print(response.output)
```

```python
timestamp = datetime.datetime.now()
```

```python
with open(f"Output/RET_{timestamp.strftime("%Y%m%d_%H%M%S")}.txt",'w') as fh:
    fh.write(response.output_text)
    
```
