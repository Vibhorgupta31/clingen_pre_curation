```python
# Library imports
import datetime
from dotenv import load_dotenv
from glob import glob
from openai import OpenAI
import os
import requests
```

```python
load_dotenv()
```

```python
try:
    with open("Prompts/prompt_v3.txt","r") as fh:
        system_prompt = fh.read()
except FileNotFoundError:
    print("Prompts/prompt_v3.txt not found")
    system_prompt = ''
```

```python
# Model parameters for reference
model = 'gpt-5'
reasoning={"effort":"low"}
```

```python
gene_info_files = sorted(glob("Input/*.txt"))
len(gene_info_files)
```

```python
### Genes for POC experiment
- LUMP: RET, FHL1, SCN8A, BPTF, CAV3, CRYAB
- SPLIT : RET, FH, SCN8A, CAV3, MSH2
- Uncertain: CRYAB 
```

```python
for gene_info_file in gene_info_files:
    try:
        with open(gene_info_file,"r") as fh:
            user_input = fh.read()
            gene = os.path.basename(gene_info_files[0]).strip('.txt')
        try:
            client = OpenAI()
            response = client.responses.create(
                model="gpt-5",
                reasoning={"effort":"low"},
                input=[
                    {
                        "role":"system",
                        "content":system_prompt
                    },
                    {
                        "role":"user",
                        "content":user_input
                    }
                ]
            )
            print(f"{os.path.basename(gene_info_files[0])} - {response.status}")
        except Exception as e:
            print(e)
            continue
        timestamp = datetime.datetime.now()
        with open(f"Output/{gene}_{timestamp.strftime("%Y%m%d_%H%M%S")}.json",'w') as fh:
            fh.write(response.output_text)
    except FileNotFoundError:
        print(f"{os.path.basename(gene_info_files[0])} not found")
        user_input = ''
```
